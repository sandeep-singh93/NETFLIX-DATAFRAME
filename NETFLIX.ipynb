### 4.In which year highest number of tv shows and movies were released? 


######  i) To answer that lets just create a "year" column with datatype "int".

```python
## Now this will convert datatype of "Release_Date" i.e. "object"  to datetime.

data["Release_Date"]= pd.to_datetime(data["Release_Date"])
```

```python
## This will create a new column "year" in our database with "float" datatype.
data.insert(7,"year",data["Release_Date"].dt.year)
```

```python
## now it will check all the rows with null values.

data[data["year"].isna()]
```

```python
### it will replace all null values with 0

data["year"].fillna(0,inplace=True)
```

```python
### it will convert datatype of "year" to "int".

data["year"]=data["year"].astype("int")
```

```python
data.dtypes                 ### it has converted datatype to int.
```

```python
### it will count all the values in "Category" column and group the data by "year"

data.groupby("year")["Category"].count().sort_values(ascending=False) 
```
